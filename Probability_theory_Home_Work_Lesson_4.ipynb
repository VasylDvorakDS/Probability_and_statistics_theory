**Задача 1.**

1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

Среднее значение и дисперсия для равномерно распределённой величины расчитывается по формулам:


image.png



```python
def midle_evently(a, b):
  return (a+b)/2
```

```python
def disp_evently(a, b):
  return (b-a)**2/12
```

```python
print('Среднее значение=',midle_evently(200, 800))
print('Дисперсия=',disp_evently(200, 800))
```

Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.


В Z сигмах лежит то или иное значение случайной величины от среднего арифметического.

image.png

Используем функцию


import scipy.stats

scipy.stats.norm.sf(abs(z))



для вычисления p-значения из z-score, вместо обращения к таблице.

```python
import scipy.stats
def prob(X, m, std):
  Z=(X-m)/std
  return scipy.stats.norm.sf(abs(Z))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см: ', prob(182, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см: ', prob(190, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см: ', 1-prob(166, 174, 8)-prob(190, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см: ', 1-prob(166, 174, 8)-prob(182, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см: ', 1-prob(158, 174, 8)-prob(190, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см: ', prob(150, 174, 8)+prob(190, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см: ', prob(150,174, 8)+prob(198, 174, 8))
```

```python
print('Вероятность того, что случайным образом выбранный взрослый человек имеет рост ниже 166 см.: ', prob(166, 174, 8))
```

**Задача 5**

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?


В Z сигмах лежит то или иное значение случайной величины от среднего арифметического.

image.png

```python
X=190
M=178
D=25

Z = (X-M)/math.sqrt(D)
print('Ответ: на', Z,' сигма.' )
```
